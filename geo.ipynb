```python
import glob
import pandas as pd

from epitator.annotator import AnnoDoc
from epitator.geoname_annotator import GeonameAnnotator
from tqdm import tqdm
```

```python
text_dir = "allpapers"
```

```python
fns = []
texts = []
for fn in glob.glob(text_dir + "/*.txt"):
    fns.append(fn.split("/")[1])
    with open(fn, 'r') as f:
        texts.append(f.read())
```

```python
doc = AnnoDoc(texts[0])
doc
```

```python
doc.add_tiers(GeonameAnnotator())
```

```python
geo_annos = doc.tiers["geonames"].spans
```

```python
geo_annos
```

```python
for anno in geo_annos:
    print(anno.geoname["name"], anno.geoname["latitude"], anno.geoname["longitude"])
```

```python
place_names = []
place_lats = []
place_longs = []
place_fns = []
```

```python
for i, text in enumerate(tqdm(texts)):
        doc = AnnoDoc(text)
        doc.add_tiers(GeonameAnnotator())
        annos = doc.tiers["geonames"].spans
        for anno in annos:        
            place_names.append(anno.geoname["name"])
            place_lats.append(anno.geoname["latitude"])
            place_longs.append(anno.geoname["longitude"])
            place_fns.append(fns[i])
```

```python
data = {
    "filename": place_fns,
    "place_name": place_names,
    "latitude": place_lats,
    "longitude": place_longs
}
data
```

```python

df = pd.DataFrame(data = data)
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
ent_df = pd.read_csv("entities.csv")
ent_df
```

```python
df_full = pd.merge(df, ent_df, on="filename")
```

```python
def wgs84_to_web_mercator(df, lon="longitude", lat="latitude"):
    """Converts decimal longitude/latitude to Web Mercator format"""
    k = 6378137
    df["x"] = df[lon] * (k * np.pi/180.0)
    df["y"] = np.log(np.tan((90 + df[lat]) * np.pi/360.0)) * k
    return df
```
